```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
#Barplo adn countplots
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

CountPlot

```python
sns.countplot(x='sex',data=tips)
```

Boxplot and Violinplot

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
#Can do entire dataframe with orient='h'
sns.boxplot(data=tips,orient='h',palette='rainbow')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker', palette="coolwarm")
```

violinplot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',hue='sex',data=tips)
```

```python
sns.violinplot(x='sex',y='total_bill',hue='smoker',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

Stripplot and Swarmplot

```python
sns.stripplot(x='day',y='total_bill',hue='smoker',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
#Swarmplot
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',hue='sex',data=tips,split=True)
```

```python
#Combining Categorical plots
```

```python
sns.violinplot(x='tip',y='day',data=tips)
sns.swarmplot(x='tip',y='day',data=tips,color='black',size=3)
```

```python
#Factor plot
```

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind='bar')
```
